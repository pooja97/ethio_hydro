#### Using Geopy package to get the country code from the lat long values

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
def reverseGeocode(coordinates): 
    result = rg.search(coordinates) 
    return result[-1]
```

```python
reverseGeocode((12.75, 41.75))
```

```python
from opencage.geocoder import OpenCageGeocode
key = "6f43823ea23f4a118b14d896b8e13ed5"
geocoder = OpenCageGeocode(key)

def reverseGeoCoder(lat,long):
    result = geocoder.reverse_geocode(lat,long)
    return (result[0]['components']['country_code'])
```

```python
reverseGeoCoder(12.75,41.75)
```

```python
def locationFilter(lat_long):
    location = geolocator.reverse(lat_long,timeout=None,language='en')
    address = location.raw['address']
    return address
```

### Finding the Monthly mean temperature 

```python
temperature_monthly_df = result.copy()
```

#### Finding the nearest neighbors. (Haversine function for finding the closest lat long)

```python
date_split_df['lat_radian'],date_split_df['long_radian'] = np.radians(date_split_df['lat']),np.radians(date_split_df['long'])
```

```python
date_split_df.drop('State',axis=1,inplace = True)
```

```python
# date_split_df

date_split_df['dLON'] = date_split_df['long_radian'] - math.radians(37.25463867187501)
date_split_df['dLAT'] = date_split_df['lat_radian'] - math.radians(9.427386615032363)
date_split_df['distance'] = 6371 * 2 * np.arcsin(np.sqrt(np.sin(date_split_df['dLAT']/2)**2 + math.cos(math.radians(9.427386615032363)) * np.cos(date_split_df['lat_radian']) * np.sin(date_split_df['dLON']/2)**2))
```

### Annual average temperature

#### Annual average temperature using multiprocessing and pool 

```python
annual_avg_df = date_split_df.copy()
annual_avg_df.head()
```

```python
#returns the list of unique year values
year_lst = annual_avg_df['Year'].unique()
```

```python
year_lst
```

```python
annual_avg_dataframe = annual_avg_df.groupby(['lat_long','Year'])
```

```python
def annual_avg(df,lat_long_lst,year_lst):
    annual_avg_df_lst = list()
    for i in lat_long_lst:
        for j in year_lst:
            d = df.get_group((i,j))
            d['Annual avg'] = (d['daily_avg'].sum())/len(d)
            annual_avg_df_lst.append(d)
    return annual_avg_df_lst
```

```python
annual_avg_result = annual_avg(annual_avg_dataframe,lat_long_list,year_lst)
```

```python
annual_average = pd.concat(annual_avg_result,axis = 0)
```
