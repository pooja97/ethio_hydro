```python
def date_split_func(lat_long_df_precipitation_1,precipitation_1_df):
    date_split_precp_1 = list()
    for i in lat_long_df_precipitation_1:
        df = precipitation_1_df.get_group(i)
        df[['Year','Month','Day']] = df['date'].str.split('-',expand = True)
        date_split_precp_1.append(df)
    return date_split_precp_1
```

```python
result = date_split_func(lat_long_df_precipitation_1,precipitation_1_df)
```
